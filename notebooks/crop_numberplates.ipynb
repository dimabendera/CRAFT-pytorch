```python
import cv2
import os
import sys

import matplotlib.pyplot as plt
```

```python
bboxes = {
    "1.jpeg": [[44.0, 479.0, 94.0, 573.0, 0.9326172, 0.0]],
    "11.jpeg": [[1129.0, 293.0, 1211.0, 381.0, 0.9370117, 0.0]],
}
```

```python
for root, dirs, files in os.walk("../data/images", topdown=False):
    for name in files:
        img = cv2.imread(os.path.join(root, name))
        basename = os.path.basename(name)
        bbox = bboxes[basename]
        print(basename, bbox)
        cropped_np = img[int(bbox[0][1]):int(bbox[0][3]), int(bbox[0][0]):int(bbox[0][2])]
        plt.imshow(cropped_np)
        plt.show()
        cv2.imwrite(os.path.join("../data/cropped_images", basename), cropped_np)
```
